```python
f = open('attrs_hk_raw.txt','r',encoding='utf-8')
f1 = open('node_type.txt','w',encoding='utf-8')
for i,line in enumerate(f):
    line = line.strip().split()
    for attr in line:
        attr = str.lower(attr)
    if 'group_related_cases_None' in line and 'classification_local' in line:
        f1.write(str(i+1)+' '+'classification_local2'+'\n')
        continue
    
    if 'group_related_cases_None' in line and 'classification_local_possibly' in line:
        f1.write(str(i+1)+' '+'classification_local2'+'\n')
        continue
        
    for attr in line:
        
        if 'classification' in attr and 'zh' not in attr:
            f1.write(str(i+1)+' '+attr+'\n')
f1.close()
f.close()
```

```python
f = open('attrs_hk_raw.txt','r',encoding='utf-8')
types = {}
count = 0
for i,line in enumerate(f):
    line = line.strip().split()
    if 'group_related_cases_None' in line:
        if 'classification_local_possibly' in line:
            print(line)
        
        if 'classification_imported_close_contact' in line or 'classification_local_close_contact' in line:
#             print(line)
            count += 1
        
        for attr in line:
            if 'classification' in attr and 'zh' not in attr:
                attr = str.lower(attr)
                if attr not in types:
                    types[attr] = 1
                else:
                    types[attr] += 1
            
            
print(types)
print(count)
f.close()
```
